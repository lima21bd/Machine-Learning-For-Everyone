```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
boston = pd.read_csv('boston.csv')
print(boston.keys())
print(boston.head())
```

```python
X = boston.drop('medv', axis = 1).values
y = boston['medv'].values
```

```python
X_rooms = X[ :,5]

print('Colum - chas >>', X_rooms)
```

```python
print(type(X_rooms))
print(type(y))
```

```python
X_rooms = X_rooms.reshape(-1, 1)
y = y.reshape(-1, 1)

print(X_rooms)
print(y)
```

Documentation : https://numpy.org/doc/stable/reference/generated/numpy.reshape.html

```python
plt.scatter(X_rooms, y)
plt.xlabel('Number of Rooms')
plt.ylabel('Values of house in $')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_rooms,  y)


prediction_space = np.linspace(min(X_rooms), max(X_rooms)).reshape(-1, 1)

print(prediction_space )
```

Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
plt.scatter(X_rooms, y, color = 'yellow')
plt.plot (prediction_space, reg.predict(prediction_space), color = 'blue')
plt.show()
```
